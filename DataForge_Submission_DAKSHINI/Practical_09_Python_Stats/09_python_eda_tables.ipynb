```python
import pandas as pd
import scipy.stats as stats
import glob
import os

# 1. Search for any CSV file in the current folder, parent folder, or data folder
search_paths = ['*.csv', '../*.csv', '../../*.csv', '../data/*.csv']
csv_files = []

for path in search_paths:
    csv_files.extend(glob.glob(path))

# 2. Safely load the first CSV it finds
if csv_files:
    dataset_path = csv_files[0]
    print(f"✅ SUCCESS: Found and loaded dataset at -> {dataset_path}")
    df = pd.read_csv(dataset_path)
else:
    print("❌ ERROR: No CSV file found.")
    print(f"Your notebook is currently looking inside: {os.getcwd()}")
    print("Please ensure your dataset (.csv) is saved inside the '04_python' folder.")
```

```python
# Summary statistics
numeric_summary = df.describe().T
numeric_summary.to_csv("../outputs/tables/numeric_summary.csv")

# Missing values
missing = (df.isna().mean() * 100).sort_values(ascending=False)
missing.to_csv("../outputs/tables/missing_percent.csv")
print("✅ Cell 2 Complete")
```

```python
# Filter numeric columns and correlate
numeric_df = df.select_dtypes(include='number')
corr_matrix = numeric_df.corr()
corr_matrix.to_csv("../outputs/tables/correlation_matrix.csv")

# Identify top correlation (ignoring 1.0 self-correlation)
top_corr = corr_matrix.unstack().sort_values(ascending=False)
top_corr = top_corr[top_corr < 1.0].head(1)
print("Highest Positive Correlation:\n", top_corr)

# Skewness (Checking if 'Sales' exists first to prevent errors)
if 'Sales' in df.columns:
    print("Sales Skewness:\n", stats.skew(df['Sales'].dropna()))
else:
    print("No column named 'Sales' found in this dataset.")
print("✅ Cell 3 Complete")
```
