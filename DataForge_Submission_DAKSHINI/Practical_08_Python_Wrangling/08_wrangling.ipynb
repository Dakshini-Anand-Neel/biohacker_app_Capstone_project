```python
import pandas as pd
import sys
import os
from pathlib import Path

# --- 1. BYPASS VS CODE PATH ISSUES ---
current_dir = os.getcwd()
sys.path.extend([
    os.path.join(current_dir, '04_python', 'scripts'),
    os.path.abspath(os.path.join(current_dir, '..', 'scripts'))
])

from clean_dataset import clean_data
print("✅ Script imported successfully!")

# --- 2. LOAD & PRE-CLEAN THE DATA ---
# (Replace with your actual file path. If it fails, try '01_data/raw/your_file.csv')
df = pd.read_csv('../../01_data/raw/dataset_raw.csv')

# Force the Sales column to be a number (strips out non-numeric characters)
if 'Sales' in df.columns:
    df['Sales'] = df['Sales'].astype(str).str.replace(r'[^0-9.-]', '', regex=True)
    df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')

# --- 3. RUN THE CLEANER & SAVE MAIN FILE ---
df_cleaned = clean_data(df)
df_cleaned.to_csv('../../01_data/processed/dataforge_cleaned.csv', index=False)

# --- 4. AGGREGATE & FORCE-SAVE TO CURRENT FOLDER ---
df_agg = df_cleaned.groupby('Region')['Sales'].mean().reset_index()

cwd = Path.cwd()
notebook_folder = cwd
for candidate in [cwd, cwd / '04_python' / 'notebooks', cwd / 'notebooks', cwd.parent / '04_python' / 'notebooks']:
    if candidate.is_dir() and candidate.name == 'notebooks':
        notebook_folder = candidate
        break

exact_file_path = notebook_folder / 'region_sales_aggregated.csv'
df_agg.to_csv(str(exact_file_path), index=False)

print(f"✅ Aggregation saved exactly here:\n{exact_file_path}")

# Pop open the folder automatically so you can see it!
try:
    os.startfile(str(notebook_folder))
except:
    pass
```
