```python
import pandas as pd

# Notice the double .csv.csv at the very end!
df = pd.read_csv(r"E:\DataForge_Internship_DAKSHINI\DataForge_EDA_Sprint_2026_2026CS001\01_data\raw\dataset_raw.csv.csv")

# Display the first few rows to confirm it worked
df.head()
```

```python
# Display basics
display(df.head())
display(df.tail())

# Display info and memory usage
df.info(memory_usage="deep")
display(df.describe())
display(df.isnull().sum())

# Display specific shape and attributes
print("Exact Shape:", df.shape)
print("Columns:", df.columns)
print("Data Types:\n", df.dtypes)
```

```python
schema = pd.DataFrame({
    "column": df.columns,
    "dtype": [str(df[c].dtype) for c in df.columns],
    "missing": [df[c].isna().sum() for c in df.columns],
    "unique": [df[c].nunique() for c in df.columns],
    "sample": [df[c].dropna().iloc[0] if df[c].dropna().shape[0] else "" for c in df.columns]
})
display(schema)
```

```python
import os

notebook_dir = r"E:\DataForge_Internship_DAKSHINI\DataForge_EDA_Sprint_2026_2026CS001\04_python\notebooks"

# Create data_info.txt
with open(rf"{notebook_dir}\data_info.txt", "w") as f:
    f.write(f"Exact Shape: {df.shape}\n")
    f.write(f"Memory Usage: {df.memory_usage(deep=True).sum()} bytes\n")
print("✅ data_info.txt created successfully!")
```
